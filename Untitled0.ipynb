```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

def task_4_3_1():
    x = np.arange(0.0, 10, 0.1)
    y = np.abs(np.sin(x*np.exp(np.cos(x))))
    plt.grid()
    plt.plot(x, y, c = "r")
    plt.fill_between(x, y)

    area = trapz(y)
    print(area)
task_4_3_1()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps
from numpy import trapz

data_list = np.arange(1, 10, 1)
plot_function = np.abs(np.cos(data_list * np.exp(np.cos(data_list) + np.log(data_list + 1))))

area = trapz(plot_function)
print("area of this plot is ", area)

plt.grid()
plt.plot([i for i in range(len(data_list))], data_list)
```

```python
import math

def calculator():
    print("Welcome to the math calculator!")
    print("Operations: +, -, *, /, e^(x+y), sin(x+y), cos(x+y), x^y")


    x = float(input("Enter the number x: "))
    y = float(input("Enter the number y: "))


    operation = input("Enter the operation (+, -, *, /, exp, sin, cos, pow): ")


    if operation == '+':
        result = x + y
    elif operation == '-':
        result = x - y
    elif operation == '*':
        result = x * y
    elif operation == '/':
        if y == 0:
            print("Error division by zero!")
            return
        result = x / y
    elif operation == 'exp':
        result = math.exp(x + y)
    elif operation == 'sin':
        result = math.sin(x + y)
    elif operation == 'cos':
        result = math.cos(x + y)
    elif operation == 'pow':
        result = x ** y
    else:
        print("Error: idk what heppened!")
        return


    print(f"Result: {result}")


calculator()
```
